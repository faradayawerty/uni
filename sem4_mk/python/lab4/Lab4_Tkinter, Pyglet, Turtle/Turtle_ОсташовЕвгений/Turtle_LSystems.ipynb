# Turtle. Фракталы

## Построение фракталов с помощью L-систем

### Кривая Коха

```python
# алфавит: F + -
# переменные: F
# константы: + -
# старт: F
# правила: (F = F+F-F-F+F)

import turtle

screen = turtle.Screen()
turtle.tracer(0)
t = turtle.Turtle()

t.hideturtle()
t.penup()
t.goto(-250, 250)
t.pendown()

axiom = 'F' # начальная строка
axmTemp = '' # временная строка для преобразования начальной
rules = {'F':'F+F-F-F+F', '+':'+', '-':'-'} # правила изменения строки

length = 20 # длина фрагмента
angle = 90
n = 0 # количество итераций

# преобразуем начальную строку
for i in range(n):
    for ch in axiom:
        axmTemp += rules[ch]
    axiom = axmTemp
    axmTemp = ''

# построим кривую в соответствии с получившейся строкой
for ch in axiom:
    if ch == '+':
        t.right(angle)
    elif ch == '-':
        t.left(angle)
    elif ch == 'F':
        t.forward(length)

turtle.tracer(1)
turtle.exitonclick()
turtle.done()
```

### Треугольник Серпинского

```python
# алфавит: F G + -
# переменные: F G
# константы: + -
# старт: F-G-G
# правила: (F = F-G+F+G-F), (G = GG)

import turtle

screen = turtle.Screen()
screen.bgcolor('black')
turtle.tracer(0)
t = turtle.Turtle()
t.color('yellow')

t.hideturtle()
t.penup()
t.goto(-320, 320)
t.setheading(-90)
t.pendown()

axiom = 'F-G-G' # начальная строка
axmTemp = '' # временная строка для преобразования начальной
rules = {'F':'F-G+F+G-F', 'G':'GG', '+':'+', '-':'-'} # правила изменения строки

length = 39 # длина фрагмента
angle = 120
n = 4 # количество итераций

# преобразуем начальную строку
for i in range(n):
    for ch in axiom:
        axmTemp += rules[ch]
    axiom = axmTemp
    axmTemp = ''

# построим кривую в соответствии с получившейся строкой
for ch in axiom:
    if ch == '+':
        t.right(angle)
    elif ch == '-':
        t.left(angle)
    elif ch == 'F':
        t.forward(length)
    elif ch == 'G':
        t.forward(length)

turtle.tracer(1)
turtle.exitonclick()
turtle.done()
```

### Стреловидная кривая Серпинского

```python
# алфавит: A B + -
# переменные: A B
# константы: + -
# старт: A
# правила: (A = B-A-B), (B = A+B+A)

import turtle

screen = turtle.Screen()
turtle.tracer(0)
t = turtle.Turtle()

t.hideturtle()
t.penup()
t.goto(-350, -300)
#t.setheading(-90)
t.pendown()

axiom = 'A' # начальная строка
axmTemp = '' # временная строка для преобразования начальной
rules = {'A':'B-A-B', 'B':'A+B+A', '+':'+', '-':'-'} # правила изменения строки

length = 5 # длина фрагмента
angle = 60
n = 8 # количество итераций

# преобразуем начальную строку
for i in range(n):
    for ch in axiom:
        axmTemp += rules[ch]
    axiom = axmTemp
    axmTemp = ''

# построим кривую в соответствии с получившейся строкой
for ch in axiom:
    if ch == '+':
        t.left(angle)
    elif ch == '-':
        t.right(angle)
    elif ch == 'A':
        t.forward(length)
    elif ch == 'B':
        t.forward(length)

turtle.tracer(1)
turtle.exitonclick()
turtle.done()
```

### Дерево Пифагора

```python
# алфавит: 0 1 [ ]
# переменные: 0 1
# константы: [ ]
# старт: 0
# правила: (1 = 11), (0 = 1[0]0)

import turtle

screen = turtle.Screen()
turtle.tracer(0)
t = turtle.Turtle()

t.hideturtle()
t.penup()
t.goto(0, -300)
t.setheading(90)
t.pendown()

axiom = '0' # начальная строка
axmTemp = '' # временная строка для преобразования начальной
rules = {'0':'1[0]0', '1':'11', '[':'[', ']':']'} # правила изменения строки

length = 10 # длина фрагмента
angle = 45
n = 4 # количество итераций
stack = []

# преобразуем начальную строку
for i in range(n):
    for ch in axiom:
        axmTemp += rules[ch]
    axiom = axmTemp
    axmTemp = ''

# построим кривую в соответствии с получившейся строкой
for ch in axiom:
    if ch == '0':
        t.forward(length)
    elif ch == '1':
        t.forward(length)
    elif ch == '[':
        stack.append(t.xcor())
        stack.append(t.ycor())
        stack.append(t.heading())
        t.left(angle)
    elif ch == ']':
        t.penup()
        t.setheading(stack.pop())
        t.sety(stack.pop())
        t.setx(stack.pop())
        t.pendown()
        t.right(angle)

turtle.tracer(1)
turtle.exitonclick()
turtle.done()
```

### Более реалистичные деревья

```python
# алфавит: 0 1 2 [ ]
# переменные: 0 1
# константы: 2 [ ]
# старт: 222220
# правила: (1 = 11), (0 = 1[0]0)

import turtle, random

screen = turtle.Screen()
screen.bgcolor('#FAFAD2')
turtle.setup(1000, 700)
turtle.tracer(0)
t = turtle.Turtle()

t.hideturtle()
t.penup()
t.goto(0, -300)
t.setheading(90)
t.pendown()

axiom = '222220' # начальная строка
axmTemp = '' # временная строка для преобразования начальной
rules = {'0':'1[20]20', '1':'21', '2':'2', '[':'[', ']':']'} # правила изменения строки

length = 10 # длина фрагмента
angle = 16
n = 12 # количество итераций
stack = []
thick = 16

# преобразуем начальную строку
for i in range(n):
    for ch in axiom:
        axmTemp += rules[ch]
    axiom = axmTemp
    axmTemp = ''

t.width(thick)
t.color('#8B4513')
# построим кривую в соответствии с получившейся строкой
for ch in axiom:
    if ch == '0':
        stack.append(t.width())
        t.width(4)
        color = random.randint(0, 10)
        if color < 3:
            t.color("#ADFF2F")
        elif color > 6:
            t.color("#98FB98")
        else:
            t.color("#3CB371")
        t.forward(length - 2)
        t.width(stack.pop())
        t.color('#8B4513')
    elif ch == '1':
        if random.randint(0, 10) > 4:
            t.forward(length)
    elif ch == '2':
        if random.randint(0, 10) > 4:
            t.forward(length)
    elif ch == '[':
        thick *= 0.75
        t.width(thick)
        stack.append(thick)
        stack.append(t.xcor())
        stack.append(t.ycor())
        stack.append(t.heading())
        t.left(angle - random.randint(-15, 15))
    elif ch == ']':
        t.penup()
        t.setheading(stack.pop())
        t.sety(stack.pop())
        t.setx(stack.pop())
        thick = stack.pop()
        t.width(thick)
        t.right(angle - random.randint(-15, 15))
        t.pendown()

turtle.tracer(1)
turtle.exitonclick()
turtle.done()
```
